```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# seaborn разрешён (удобно для EDA)
import seaborn as sns

from pathlib import Path

# sklearn разрешён ТОЛЬКО для сравнения и для не-обязательных частей (OHE/Poly)
from sklearn.preprocessing import PolynomialFeatures, OneHotEncoder
from sklearn.linear_model import LinearRegression as SkLinearRegression
from sklearn.metrics import mean_squared_error as sk_mse, mean_absolute_error as sk_mae, r2_score as sk_r2
try:
    from sklearn.metrics import mean_absolute_percentage_error as sk_mape
    _HAS_SK_MAPE = True
except Exception:
    _HAS_SK_MAPE = False

# ----------------------------
# 0) Reproducibility
# ----------------------------
SEED = 42
rng = np.random.default_rng(SEED)

# ----------------------------
# 1) Load data
# ----------------------------
DATA_DIR = Path("data")
train = pd.read_csv(DATA_DIR / "train.csv")
test  = pd.read_csv(DATA_DIR / "test.csv")
ex    = pd.read_csv(DATA_DIR / "ex.csv")

print("Train shape:", train.shape)
print("Test shape: ", test.shape)
print("Ex shape:   ", ex.shape)
display(train.head())
display(test.head())
display(ex.head())

TARGET_COL = "RiskScore"
ID_COL = "ID"

# ----------------------------
# 2) EDA (обязательно)
# ----------------------------
train_eda = train.copy()
train_eda = train_eda[train_eda[TARGET_COL].notna()].copy()

# лёгкая "очистка" как в вашем решении (чтобы графики были адекватнее)
train_eda = train_eda[train_eda[TARGET_COL].abs() <= 200].copy()
train_eda[TARGET_COL] = train_eda[TARGET_COL].clip(0.0, 100.0)

y_eda = train_eda[TARGET_COL].astype(float)

plt.figure(figsize=(8, 4))
sns.histplot(y_eda, kde=True, bins=30)
plt.title("Распределение целевой переменной RiskScore")
plt.xlabel("RiskScore")
plt.ylabel("Count")
plt.show()

# зависимости "нескольких признаков" от таргета:
# выберем топ-3 числовых признака по |corr| с таргетом (на сырых данных)
num_cols = train_eda.select_dtypes(include=[np.number]).columns.tolist()
num_cols = [c for c in num_cols if c not in [TARGET_COL] and c != ID_COL]
corrs = train_eda[num_cols + [TARGET_COL]].corr(numeric_only=True)[TARGET_COL].drop(TARGET_COL).abs().sort_values(ascending=False)
top_feats = corrs.head(3).index.tolist()

fig, axes = plt.subplots(1, len(top_feats), figsize=(5*len(top_feats), 4))
if len(top_feats) == 1:
    axes = [axes]
for ax, f in zip(axes, top_feats):
    ax.scatter(train_eda[f], train_eda[TARGET_COL], s=10, alpha=0.5)
    ax.set_title(f"{f} vs {TARGET_COL}\n|corr|={corrs[f]:.3f}")
    ax.set_xlabel(f)
    ax.set_ylabel(TARGET_COL)
plt.tight_layout()
plt.show()

# матрица корреляций (числовые + таргет)
corr_mat = train_eda[num_cols + [TARGET_COL]].corr(numeric_only=True)
plt.figure(figsize=(10, 7))
sns.heatmap(corr_mat, cmap="coolwarm", center=0, linewidths=0.2)
plt.title("Матрица корреляций (числовые признаки + RiskScore)")
plt.show()

# краткие выводы (обязательно)
skew = float(y_eda.skew())
desc = y_eda.describe()
print("\nEDA выводы (кратко):")
print(f"- Target RiskScore: min={desc['min']:.2f}, max={desc['max']:.2f}, mean={desc['mean']:.2f}, std={desc['std']:.2f}, skew={skew:.2f}")
if top_feats:
    print(f"- Наиболее коррелирующие (|corr|) числовые признаки с RiskScore: {', '.join(top_feats)}")
print("- Корреляционная матрица показывает пары признаков с высокой линейной зависимостью → возможна мультиколлинеарность.")
print("- Для линейной регрессии важно корректное масштабирование и аккуратная работа с выбросами/лог-трансформами.\n")

# ----------------------------
# 3) Feature engineering (ваш change_features)
# ----------------------------
def change_features(df: pd.DataFrame):
    """Feature engineering как в вашем коде (inplace)."""
    if "ApplicationDate" in df.columns:
        df.drop(columns=["ApplicationDate"], inplace=True)

    mapping = {
        "High School": 1, "Associate": 2, "Bachelor": 3, "Master": 4, "Doctorate": 5,
    }
    if "EducationLevel" in df.columns:
        df["EducationLevel"] = df["EducationLevel"].map(mapping).fillna(0).astype(float)

    # агрегированный доход
    if {"MonthlyIncome","AnnualIncome"}.issubset(df.columns):
        df["Income"] = 1 / (np.log1p((df["MonthlyIncome"] + df["AnnualIncome"] / 12) / 2) + 1)

    # ставки
    if {"BaseInterestRate","InterestRate"}.issubset(df.columns):
        df["InterestRate2"] = (df["BaseInterestRate"] + df["InterestRate"]) / 2
        df["InterestRateSpread"] = df["InterestRate"] - df["BaseInterestRate"]

    # возраст – стаж
    if {"Age","Experience"}.issubset(df.columns):
        df["AgeExperience"] = df["Age"] - df["Experience"]

    # отношения и разности
    if {"NetWorth","AnnualIncome"}.issubset(df.columns):
        df["NetIncome"] = df["NetWorth"] / (df["AnnualIncome"] + 1)
        df["NetWorth"] = np.log1p(df["NetWorth"]) ** 3

    if {"LoanAmount","AnnualIncome"}.issubset(df.columns):
        df["LoanIncome"] = df["LoanAmount"] / (df["AnnualIncome"] + 1)

    if {"LengthOfCreditHistory","Age"}.issubset(df.columns):
        df["HistToAge"] = df["LengthOfCreditHistory"] / (df["Age"] + 1)

    if {"MonthlyLoanPayment","MonthlyDebtPayments","MonthlyIncome"}.issubset(df.columns):
        total = df["MonthlyLoanPayment"] + df["MonthlyDebtPayments"]
        df["DebtToIncomeRatio2"] = total / (df["MonthlyIncome"] + 1)
        df["RemainingIncome"] = df["MonthlyIncome"] - total

    if {"TotalAssets","TotalLiabilities"}.issubset(df.columns):
        df["AssetsLiabilities"] = df["TotalAssets"] / (df["TotalLiabilities"] + 1)
        df["LiabilityAssets"] = df["TotalLiabilities"] - df["TotalAssets"]

    if {"NetWorth","TotalLiabilities"}.issubset(df.columns):
        df["NetLiabilities"] = df["NetWorth"] / (df["TotalLiabilities"] + 1)

    if {"AnnualIncome","NumberOfDependents"}.issubset(df.columns):
        df["IncomeDependents"] = df["AnnualIncome"] / (df["NumberOfDependents"] + 1)

    if {"Experience","Age"}.issubset(df.columns):
        df["ExpAge"] = df["Experience"] / (df["Age"] + 1)

    # доп. лог-преобразования как в solution
    if "AnnualIncome" in df.columns:
        df["AnnualIncome"] = -np.log1p(df["AnnualIncome"])
    if "BaseInterestRate" in df.columns:
        df["BaseInterestRate"] = np.log1p(df["BaseInterestRate"])
    if "LoanAmount" in df.columns:
        df["LoanAmount"] = -np.log1p(df["LoanAmount"])

    # биннинг CreditScore
    if "CreditScore" in df.columns:
        scores = [300, 500, 600, 650, 700, 750, 800, 850, 900]
        names = [f"{scores[i]}-{scores[i+1]}" for i in range(len(scores) - 1)]
        df["BinsCreditScore"] = pd.cut(df["CreditScore"], bins=scores, labels=names, include_lowest=True).astype(str)

# ----------------------------
# 4) Manual scalers (обязательно): Z-Score и Min-Max
# ----------------------------
class ZScoreScaler:
    def __init__(self, eps=1e-12):
        self.eps = eps
        self.mean_ = None
        self.std_ = None

    def fit(self, X):
        X = np.asarray(X, dtype=float)
        self.mean_ = X.mean(axis=0)
        self.std_ = X.std(axis=0)
        self.std_ = np.where(self.std_ < self.eps, 1.0, self.std_)
        return self

    def transform(self, X):
        X = np.asarray(X, dtype=float)
        return (X - self.mean_) / self.std_

    def fit_transform(self, X):
        return self.fit(X).transform(X)

class MinMaxScaler:
    def __init__(self, eps=1e-12):
        self.eps = eps
        self.min_ = None
        self.max_ = None

    def fit(self, X):
        X = np.asarray(X, dtype=float)
        self.min_ = X.min(axis=0)
        self.max_ = X.max(axis=0)
        return self

    def transform(self, X):
        X = np.asarray(X, dtype=float)
        denom = (self.max_ - self.min_)
        denom = np.where(np.abs(denom) < self.eps, 1.0, denom)
        return (X - self.min_) / denom

    def fit_transform(self, X):
        return self.fit(X).transform(X)

# ----------------------------
# 5) Preprocessor (ваша логика, но scaling теперь через наши скейлеры)
# ----------------------------
class SolutionPreprocessor:
    """
    Реплика вашего препроцессинга:
    - числовые: median -> sign*log1p(abs) -> масштабирование (z-score или minmax) [ручная реализация]
    - полиномиальные признаки degree
    - категориальные: mode -> OneHotEncoder
    - отбор признаков по модулю корреляции с таргетом (top_frac)
    """
    def __init__(self, scaling_method="zscore", polynomial_degree=2, top_frac=0.35):
        self.scaling_method = scaling_method
        self.polynomial_degree = polynomial_degree
        self.top_frac = top_frac

    @staticmethod
    def _apply_log_transform(df_num: pd.DataFrame) -> pd.DataFrame:
        return np.sign(df_num) * np.log1p(np.abs(df_num))

    def _make_ohe(self):
        try:
            return OneHotEncoder(handle_unknown="ignore", sparse_output=False)
        except TypeError:
            return OneHotEncoder(handle_unknown="ignore", sparse=False)

    def fit(self, X, y):
        if y is None:
            raise ValueError("SolutionPreprocessor требует y в fit (для отбора признаков по корреляции).")

        X = pd.DataFrame(X).reset_index(drop=True)
        y = pd.Series(y).reset_index(drop=True).astype(float)

        self.numeric_cols_ = X.select_dtypes(include=[np.number]).columns.tolist()
        self.categorical_cols_ = [c for c in X.columns if c not in self.numeric_cols_]

        # ---- numeric ----
        if self.numeric_cols_:
            num = X[self.numeric_cols_].copy()
            self.numeric_imputer_ = num.median()
            num = num.fillna(self.numeric_imputer_)
            num = self._apply_log_transform(num)

            if self.scaling_method.lower() == "zscore":
                self.num_scaler_ = ZScoreScaler().fit(num.values)
            else:
                self.num_scaler_ = MinMaxScaler().fit(num.values)
            num_scaled = self.num_scaler_.transform(num.values)

            self.poly_ = PolynomialFeatures(degree=self.polynomial_degree, include_bias=False)
            poly_arr = self.poly_.fit_transform(num_scaled)
            poly_cols = self.poly_.get_feature_names_out(self.numeric_cols_)
            poly_df = pd.DataFrame(poly_arr, columns=poly_cols, index=X.index)

            # доп. нормировка полиномов (ручная)
            self.poly_scaler_ = ZScoreScaler().fit(poly_df.values)
            poly_df = pd.DataFrame(self.poly_scaler_.transform(poly_df.values), columns=poly_cols, index=X.index)
        else:
            self.numeric_imputer_ = None
            self.num_scaler_ = None
            self.poly_ = None
            self.poly_scaler_ = None
            poly_df = pd.DataFrame(index=X.index)

        # ---- categorical ----
        if self.categorical_cols_:
            cat = X[self.categorical_cols_].copy()
            self.categorical_imputer_ = cat.mode(dropna=True).iloc[0] if len(cat.columns) else None
            if self.categorical_imputer_ is not None:
                cat = cat.fillna(self.categorical_imputer_)
            self.ohe_ = self._make_ohe()
            cat_arr = self.ohe_.fit_transform(cat)
            cat_cols = self.ohe_.get_feature_names_out(self.categorical_cols_)
            cat_df = pd.DataFrame(cat_arr, columns=cat_cols, index=X.index)
        else:
            self.categorical_imputer_ = None
            self.ohe_ = None
            cat_df = pd.DataFrame(index=X.index)

        all_features = pd.concat([poly_df, cat_df], axis=1)

        # ---- select top by |corr| ----
        corrs = all_features.apply(lambda col: col.corr(y)).fillna(0.0)
        abs_corr = corrs.abs()
        n_features = len(all_features.columns)
        n_select = max(1, int(self.top_frac * n_features))
        self.top_features_ = abs_corr.sort_values(ascending=False).head(n_select).index.tolist()

        self._all_columns_ = all_features.columns.tolist()
        return self

    def transform(self, X):
        X = pd.DataFrame(X).copy()
        # numeric
        if self.numeric_cols_:
            num = X[self.numeric_cols_].copy()
            num = num.fillna(self.numeric_imputer_)
            num = self._apply_log_transform(num)
            num_scaled = self.num_scaler_.transform(num.values)

            poly_arr = self.poly_.transform(num_scaled)
            poly_cols = self.poly_.get_feature_names_out(self.numeric_cols_)
            poly_arr2 = self.poly_scaler_.transform(poly_arr)
            poly_df = pd.DataFrame(poly_arr2, columns=poly_cols, index=X.index)
        else:
            poly_df = pd.DataFrame(index=X.index)

        # categorical
        if self.categorical_cols_ and self.ohe_ is not None:
            cat = X[self.categorical_cols_].copy()
            if self.categorical_imputer_ is not None:
                cat = cat.fillna(self.categorical_imputer_)
            cat_arr = self.ohe_.transform(cat)
            cat_cols = self.ohe_.get_feature_names_out(self.categorical_cols_)
            cat_df = pd.DataFrame(cat_arr, columns=cat_cols, index=X.index)
        else:
            cat_df = pd.DataFrame(index=X.index)

        all_features = pd.concat([poly_df, cat_df], axis=1)
        return all_features[self.top_features_].to_numpy(dtype=float)

# ----------------------------
# 6) Manual metrics (обязательно) + сравнение со sklearn.metrics
# ----------------------------
def mse(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    return float(np.mean((y_true - y_pred) ** 2))

def mae(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    return float(np.mean(np.abs(y_true - y_pred)))

def r2(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    if ss_tot == 0:
        return 0.0
    return float(1 - ss_res / ss_tot)

def mape(y_true, y_pred, eps=1e-8):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    mask = np.abs(y_true) > eps
    if mask.sum() == 0:
        return 0.0
    return float(np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100.0)

# ----------------------------
# 7) Manual Linear Regression (обязательно): normal equation + GD + SGD + регуляризации
# ----------------------------
class MyLinearRegression:
    """
    solver:
      - 'normal' : аналитическое решение (normal equation / least squares)
      - 'gd'     : градиентный спуск
      - 'sgd'    : стохастический градиентный спуск (mini-batch)

    regularization:
      - None / 'l1' / 'l2' / 'elasticnet' / 'lp'
    """
    def __init__(self,
                 solver="normal",
                 fit_intercept=True,
                 lr=0.05,
                 epochs=500,
                 batch_size=128,
                 alpha=0.0,          # сила регуляризации
                 regularization=None,
                 l1_ratio=0.5,       # для elasticnet: доля L1 (0..1)
                 p=2.0,              # для Lp
                 random_state=42,
                 tol=1e-9):
        self.solver = solver
        self.fit_intercept = fit_intercept
        self.lr = lr
        self.epochs = epochs
        self.batch_size = batch_size
        self.alpha = float(alpha)
        self.regularization = regularization
        self.l1_ratio = float(l1_ratio)
        self.p = float(p)
        self.random_state = random_state
        self.tol = tol
        self.coef_ = None
        self.intercept_ = None

    def _add_bias(self, X):
        if not self.fit_intercept:
            return X
        return np.hstack([np.ones((X.shape[0], 1), dtype=float), X])

    def _split_w(self, w):
        if self.fit_intercept:
            self.intercept_ = float(w[0])
            self.coef_ = w[1:].copy()
        else:
            self.intercept_ = 0.0
            self.coef_ = w.copy()

    def _reg_grad(self, w):
        """
        gradient of regularization term w.r.t w
        IMPORTANT: intercept (w[0]) is NOT regularized.
        """
        if self.alpha <= 0 or self.regularization is None:
            return np.zeros_like(w)

        g = np.zeros_like(w)
        ww = w.copy()
        if self.fit_intercept:
            ww[0] = 0.0  # do not regularize intercept

        reg = self.regularization.lower()
        if reg == "l2":
            g = 2.0 * self.alpha * ww
        elif reg == "l1":
            g = self.alpha * np.sign(ww)
        elif reg == "elasticnet":
            # alpha*( l1_ratio*|w| + (1-l1_ratio)*||w||^2 )
            g = self.alpha * self.l1_ratio * np.sign(ww) + 2.0 * self.alpha * (1.0 - self.l1_ratio) * ww
        elif reg == "lp":
            # alpha * sum |w|^p, grad = alpha * p * |w|^(p-1) * sign(w)
            p = max(self.p, 1e-6)
            g = self.alpha * p * (np.abs(ww) ** (p - 1.0)) * np.sign(ww)
        else:
            # неизвестный тип — без рег
            g = np.zeros_like(w)

        if self.fit_intercept:
            g[0] = 0.0
        return g

    def fit(self, X, y):
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=float).reshape(-1)

        Xb = self._add_bias(X)

        if self.solver.lower() == "normal":
            # аналитика: least squares / normal eq
            # поддержим ridge (L2) в closed-form, остальное — только GD/SGD
            if self.alpha > 0 and (self.regularization or "").lower() == "l2":
                # (X^T X + alpha*I)^{-1} X^T y, intercept не регуляризуем
                XtX = Xb.T @ Xb
                I = np.eye(XtX.shape[0])
                if self.fit_intercept:
                    I[0, 0] = 0.0
                A = XtX + self.alpha * I
                b = Xb.T @ y
                w = np.linalg.solve(A, b)
            else:
                # обычный МНК
                w, *_ = np.linalg.lstsq(Xb, y, rcond=None)
            self._split_w(w)
            return self

        # GD / SGD
        w = np.zeros(Xb.shape[1], dtype=float)
        rs = np.random.default_rng(self.random_state)

        prev_loss = np.inf
        n = Xb.shape[0]

        def _loss(y_true, y_pred):
            return np.mean((y_true - y_pred) ** 2)

        if self.solver.lower() == "gd":
            for ep in range(self.epochs):
                y_pred = Xb @ w
                grad = (2.0 / n) * (Xb.T @ (y_pred - y)) + self._reg_grad(w)
                w -= self.lr * grad

                # простая остановка
                if ep % 50 == 0 or ep == self.epochs - 1:
                    cur_loss = _loss(y, Xb @ w)
                    if abs(prev_loss - cur_loss) < self.tol:
                        break
                    prev_loss = cur_loss

        elif self.solver.lower() == "sgd":
            bs = max(1, int(self.batch_size))
            for ep in range(self.epochs):
                idx = rs.permutation(n)
                Xs = Xb[idx]
                ys = y[idx]
                for start in range(0, n, bs):
                    xb = Xs[start:start+bs]
                    yb = ys[start:start+bs]
                    y_pred = xb @ w
                    grad = (2.0 / xb.shape[0]) * (xb.T @ (y_pred - yb)) + self._reg_grad(w)
                    w -= self.lr * grad

                if ep % 50 == 0 or ep == self.epochs - 1:
                    cur_loss = _loss(y, Xb @ w)
                    if abs(prev_loss - cur_loss) < self.tol:
                        break
                    prev_loss = cur_loss
        else:
            raise ValueError("solver должен быть: 'normal', 'gd' или 'sgd'")

        self._split_w(w)
        return self

    def predict(self, X):
        X = np.asarray(X, dtype=float)
        y_pred = X @ self.coef_ + self.intercept_
        return y_pred

# ----------------------------
# 8) Manual Pipeline (чтобы CV был честным без утечки)
# ----------------------------
class ManualPipeline:
    def __init__(self, preprocessor, model):
        self.preprocessor = preprocessor
        self.model = model

    def fit(self, X, y):
        self.preprocessor.fit(X, y)
        Xp = self.preprocessor.transform(X)
        self.model.fit(Xp, y)
        return self

    def predict(self, X):
        Xp = self.preprocessor.transform(X)
        return self.model.predict(Xp)

# ----------------------------
# 9) Manual CV (обязательно): K-Fold + Leave-One-Out
# ----------------------------
def kfold_split(n, k=5, shuffle=True, random_state=42):
    idx = np.arange(n)
    if shuffle:
        rs = np.random.default_rng(random_state)
        rs.shuffle(idx)
    folds = np.array_split(idx, k)
    for i in range(k):
        val_idx = folds[i]
        train_idx = np.concatenate([folds[j] for j in range(k) if j != i])
        yield train_idx, val_idx

def loo_split(n):
    idx = np.arange(n)
    for i in range(n):
        val_idx = np.array([i])
        train_idx = np.concatenate([idx[:i], idx[i+1:]])
        yield train_idx, val_idx

def evaluate_metrics(y_true, y_pred):
    return {
        "MSE": mse(y_true, y_pred),
        "MAE": mae(y_true, y_pred),
        "R2":  r2(y_true, y_pred),
        "MAPE": mape(y_true, y_pred),
    }

def manual_cv(pipeline_factory, X, y, splitter, max_folds=None):
    X = pd.DataFrame(X).reset_index(drop=True)
    y = pd.Series(y).reset_index(drop=True)
    metrics_list = []
    for fold_i, (tr, va) in enumerate(splitter):
        if max_folds is not None and fold_i >= max_folds:
            break
        pipe = pipeline_factory()
        pipe.fit(X.iloc[tr], y.iloc[tr])
        pred = pipe.predict(X.iloc[va])
        metrics_list.append(evaluate_metrics(y.iloc[va].values, pred))
    df = pd.DataFrame(metrics_list)
    return df

# ----------------------------
# 10) Prepare cleaned + FE data
# ----------------------------
train_raw = train.copy()
train_clean = train_raw[train_raw[TARGET_COL].notna()].copy()
train_clean = train_clean[train_clean[TARGET_COL].abs() <= 200].reset_index(drop=True)
train_clean[TARGET_COL] = train_clean[TARGET_COL].clip(0.0, 100.0)

print("\nИсходный train:", train_raw.shape)
print("После очистки: ", train_clean.shape)

train_fe = train_clean.copy()
test_fe  = test.copy()

change_features(train_fe)
change_features(test_fe)

feature_cols = [c for c in train_fe.columns if c not in [TARGET_COL, ID_COL]]
X = train_fe[feature_cols]
y = train_fe[TARGET_COL].astype(float)

# ----------------------------
# 11) Сравнение моделей: manual LR (normal/gd/sgd) vs sklearn LR
# ----------------------------
# Параметры препроцессора — как в вашем пайплайне
PREP_KW = dict(scaling_method="zscore", polynomial_degree=2, top_frac=0.35)

# Hold-out split (ручной)
n = len(X)
idx = np.arange(n)
rng2 = np.random.default_rng(SEED)
rng2.shuffle(idx)
val_size = int(0.2 * n)
val_idx = idx[:val_size]
tr_idx  = idx[val_size:]

X_tr, y_tr = X.iloc[tr_idx], y.iloc[tr_idx]
X_va, y_va = X.iloc[val_idx], y.iloc[val_idx]

# sklearn pipeline (для сравнения и для максимально безопасного сабмита)
prep_sk = SolutionPreprocessor(**PREP_KW)
prep_sk.fit(X_tr, y_tr)
Xtr_p = prep_sk.transform(X_tr)
Xva_p = prep_sk.transform(X_va)

sk_model = SkLinearRegression()
sk_model.fit(Xtr_p, y_tr.values)
sk_pred = sk_model.predict(Xva_p)

# manual models on the SAME processed matrix (честное сравнение именно LR-части)
m_normal = MyLinearRegression(solver="normal")
m_gd     = MyLinearRegression(solver="gd", lr=0.05, epochs=800, alpha=0.0, regularization=None)
m_sgd    = MyLinearRegression(solver="sgd", lr=0.05, epochs=800, batch_size=256, alpha=0.0, regularization=None)

m_normal.fit(Xtr_p, y_tr.values); pred_normal = m_normal.predict(Xva_p)
m_gd.fit(Xtr_p, y_tr.values);     pred_gd     = m_gd.predict(Xva_p)
m_sgd.fit(Xtr_p, y_tr.values);    pred_sgd    = m_sgd.predict(Xva_p)

print("\nHold-out метрики (на одинаковом препроцессинге):")

def _print_block(name, yy, pp):
    met = evaluate_metrics(yy, pp)
    print(f"[{name}] MSE={met['MSE']:.4f} | MAE={met['MAE']:.4f} | R2={met['R2']:.4f} | MAPE={met['MAPE']:.2f}%")
    return met

met_sk      = _print_block("sklearn.LinearRegression", y_va.values, sk_pred)
met_normal  = _print_block("MyLinearRegression (normal)", y_va.values, pred_normal)
met_gd      = _print_block("MyLinearRegression (gd)",     y_va.values, pred_gd)
met_sgd     = _print_block("MyLinearRegression (sgd)",    y_va.values, pred_sgd)

# сравнение метрик со sklearn.metrics (обязательно)
print("\nСравнение наших метрик со sklearn.metrics (на sklearn-предикте):")
print(f"  MSE:  ours={mse(y_va, sk_pred):.6f} | sklearn={sk_mse(y_va, sk_pred):.6f}")
print(f"  MAE:  ours={mae(y_va, sk_pred):.6f} | sklearn={sk_mae(y_va, sk_pred):.6f}")
print(f"  R2:   ours={r2(y_va, sk_pred):.6f}  | sklearn={sk_r2(y_va, sk_pred):.6f}")
if _HAS_SK_MAPE:
    print(f"  MAPE: ours={mape(y_va, sk_pred):.6f} | sklearn={sk_mape(y_va, sk_pred)*100.0:.6f}")
else:
    print(f"  MAPE: ours={mape(y_va, sk_pred):.6f} | sklearn=not available")

# ----------------------------
# 12) Manual CV: K-Fold + LOO (обязательно)
# ----------------------------
def factory_manual_normal():
    return ManualPipeline(SolutionPreprocessor(**PREP_KW), MyLinearRegression(solver="normal"))

def factory_sklearn_lr():
    # sklearn-модель, но CV сплиттер/цикл — ручной
    class _SkPipe:
        def __init__(self):
            self.prep = SolutionPreprocessor(**PREP_KW)
            self.model = SkLinearRegression()
        def fit(self, X, y):
            self.prep.fit(X, y)
            Xp = self.prep.transform(X)
            self.model.fit(Xp, np.asarray(y, float))
            return self
        def predict(self, X):
            Xp = self.prep.transform(X)
            return self.model.predict(Xp)
    return _SkPipe()

print("\nManual K-Fold CV (k=5) — MyLinearRegression(normal):")
cv_df = manual_cv(factory_manual_normal, X, y, splitter=kfold_split(len(X), k=5, shuffle=True, random_state=SEED))
display(cv_df)
print("KFold mean:", cv_df.mean().to_dict())
print("KFold std: ", cv_df.std().to_dict())

# LOO может быть очень долгим — сделаем на подвыборке (реализация есть, но ограничим runtime)
LOO_MAX_N = min(400, len(X))  # можно увеличить при желании
X_loo = X.iloc[:LOO_MAX_N].copy()
y_loo = y.iloc[:LOO_MAX_N].copy()

print(f"\nManual Leave-One-Out CV — MyLinearRegression(normal) (на первых {LOO_MAX_N} объектах для скорости):")
loo_df = manual_cv(factory_manual_normal, X_loo, y_loo, splitter=loo_split(len(X_loo)))
display(loo_df.head())
print("LOO mean:", loo_df.mean().to_dict())

# ----------------------------
# 13) Bonus: регуляризации (демо — ElasticNet/L2/L1/Lp на hold-out)
# ----------------------------
print("\n[Bonus] Демонстрация регуляризаций (на hold-out, solver=gd):")
reg_models = [
    ("L2 (ridge)",      MyLinearRegression(solver="gd", lr=0.05, epochs=800, alpha=1e-3, regularization="l2")),
    ("L1 (lasso)",      MyLinearRegression(solver="gd", lr=0.05, epochs=800, alpha=1e-3, regularization="l1")),
    ("ElasticNet",      MyLinearRegression(solver="gd", lr=0.05, epochs=800, alpha=1e-3, regularization="elasticnet", l1_ratio=0.5)),
    ("Lp (p=1.5)",      MyLinearRegression(solver="gd", lr=0.05, epochs=800, alpha=1e-3, regularization="lp", p=1.5)),
]
for name, mdl in reg_models:
    mdl.fit(Xtr_p, y_tr.values)
    pr = mdl.predict(Xva_p)
    _print_block(name, y_va.values, pr)

# ----------------------------
# 14) Final training + submission
#     ВАЖНО: чтобы "не ухудшить скор", выбираем модель по hold-out MSE:
#     если manual(normal) не хуже sklearn — используем manual, иначе sklearn.
# ----------------------------
use_manual = (met_normal["MSE"] <= met_sk["MSE"] + 1e-9)

print("\nВыбор модели для финального обучения/сабмита:")
print(" - Используем manual(normal) ?" , use_manual)

# fit preprocessor on full train
prep_full = SolutionPreprocessor(**PREP_KW)
prep_full.fit(X, y)
X_full_p = prep_full.transform(X)

# fit chosen model on full processed
if use_manual:
    final_model = MyLinearRegression(solver="normal")
    final_model.fit(X_full_p, y.values)
else:
    final_model = SkLinearRegression()
    final_model.fit(X_full_p, y.values)

# predict on test
X_test = test_fe[feature_cols]
X_test_p = prep_full.transform(X_test)

test_pred = final_model.predict(X_test_p)
test_pred = np.clip(test_pred, 0.0, 100.0)

submission = pd.DataFrame({ID_COL: test_fe[ID_COL], TARGET_COL: test_pred})
submission.to_csv("out.csv", index=False)
print("\nSaved submission to out.csv")
display(submission.head())

# Дополнительно (не обязательно): сохранить manual-версию отдельно, если сабмит делали sklearn
if not use_manual:
    manual_model = MyLinearRegression(solver="normal")
    manual_model.fit(X_full_p, y.values)
    manual_pred = np.clip(manual_model.predict(X_test_p), 0.0, 100.0)
    pd.DataFrame({ID_COL: test_fe[ID_COL], TARGET_COL: manual_pred}).to_csv("out_manual.csv", index=False)
    print("Also saved out_manual.csv (manual normal equation).")
```
